```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.ensemble import RandomForestRegressor
from scipy.optimize import minimize

# Configuración de estilo visual "Pro"
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ==============================================================================
# 0. CONFIGURACIÓN Y DESCARGA (Incluyendo Benchmark Global)
# ==============================================================================
tickers_map = {
    'SPY': 'US Market', 'QQQ': 'Nasdaq', 'IWM': 'Russell 2000',
    'EFA': 'EAFE', 'EEM': 'Emerging', 'VGK': 'Europe', 'VPL': 'Asia',
    'LQD': 'Corp Bonds', 'HYG': 'High Yield', 'TLT': 'Treasury 20y', 'AGG': 'Total Bond',
    'SHY': 'Treasury 1-3y', 
    'GLD': 'Gold', 'SLV': 'Silver', 'USO': 'Oil', 
    'DJP': 'Commodities', 'VNQ': 'Real Estate'
}

download_list = list(tickers_map.keys()) + ['ACWI'] 

print("--- 0. Descargando Datos (Incluyendo MSCI ACWI como Benchmark) ---")
raw_data = yf.download(download_list, start="2010-01-01", interval="1mo", auto_adjust=True)['Close']
all_returns = np.log(raw_data / raw_data.shift(1)).dropna()

# Separamos Benchmark Global y Activos de Inversión
market_benchmark = all_returns['ACWI']
returns = all_returns[list(tickers_map.keys())]

# ==============================================================================
# STEP 1: TIME-SERIES ANALYSIS & VISUALIZACIÓN AVANZADA
# ==============================================================================
print("\n" + "="*60)
print("STEP 1: ANÁLISIS ESTADÍSTICO PROFUNDO")
print("="*60)

# A. Tabla de Estadísticas
analysis_data = []
for ticker in returns.columns:
    r = returns[ticker]
    covariance = np.cov(r, market_benchmark)[0][1]
    beta = covariance / np.var(market_benchmark)
    
    analysis_data.append({
        'Asset': ticker,
        'ADF (Stationary)': adfuller(r)[1] < 0.05,
        'Skewness': round(skew(r), 2),
        'Kurtosis': round(kurtosis(r), 2),
        'Global Beta': round(beta, 2),
        'Vol (Ann)': round(r.std() * np.sqrt(12), 2)
    })
print(pd.DataFrame(analysis_data).set_index('Asset'))

# --- GRÁFICA 1: MAPA DE CALOR DE CORRELACIONES ---
plt.figure(figsize=(14, 10))
corr_matrix = returns.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool)) # Ocultar mitad superior repetida
sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Mapa de Diversificación: ¿Están mis activos correlacionados?', fontsize=16)
plt.show()

# --- GRÁFICA 2: DISTRIBUCIÓN DE RETORNOS (FAT TAILS) ---
# Comparamos SPY (Normal) vs USO (Cola Pesada) para justificar CVaR
plt.figure(figsize=(10, 5))
sns.kdeplot(returns['SPY'], fill=True, label='S&P 500 (SPY)', color='blue')
sns.kdeplot(returns['USO'], fill=True, label='Petróleo (USO)', color='red')
plt.title('Justificación del CVaR: Colas Pesadas y Asimetría', fontsize=16)
plt.xlabel('Retorno Mensual Logarítmico')
plt.legend()
plt.show()

# ==============================================================================
# STEP 2: FORECASTING & EXPLICABILIDAD DE LA IA
# ==============================================================================
print("\n" + "="*60)
print("STEP 2: MODELO DE IA (FACTORES Y PREDICCIÓN)")
print("="*60)

forecasts = {}
feature_importance_acc = np.zeros(4) # Acumulador para ver qué importa globalmente
term_spread = raw_data['TLT'] - raw_data['SHY']
credit_spread = raw_data['LQD'] - raw_data['HYG']

for ticker in returns.columns:
    df_f = pd.DataFrame(index=returns.index)
    df_f['target'] = returns[ticker].shift(-1)
    df_f['Momentum (12m)'] = returns[ticker].rolling(12).mean()
    df_f['Volatility (3m)'] = returns[ticker].rolling(3).std()
    df_f['Term Spread'] = term_spread.reindex(df_f.index).ffill()
    df_f['Credit Spread'] = credit_spread.reindex(df_f.index).ffill()
    df_f = df_f.dropna()
    
    X = df_f[['Momentum (12m)', 'Volatility (3m)', 'Term Spread', 'Credit Spread']]
    y = df_f['target']
    
    model = RandomForestRegressor(n_estimators=100, max_depth=3, random_state=42)
    model.fit(X, y)
    
    # Guardamos importancia de features
    feature_importance_acc += model.feature_importances_
    
    last_features = X.iloc[[-1]]
    forecasts[ticker] = model.predict(last_features)[0]

# --- GRÁFICA 3: ¿QUÉ MIRA LA IA? (FEATURE IMPORTANCE) ---
features_names = ['Momentum (Tendencia)', 'Volatilidad (Riesgo)', 'Term Spread (Macro)', 'Credit Spread (Miedo)']
avg_importance = feature_importance_acc / len(returns.columns)
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_importance, y=features_names, palette='viridis')
plt.title('Explicabilidad del Modelo: ¿Qué factores mueven el mercado hoy?', fontsize=16)
plt.xlabel('Importancia Relativa en el Modelo')
plt.show()

expected_returns = pd.Series(forecasts)

# ==============================================================================
# STEP 3: OPTIMIZACIÓN CVaR & VISUALIZACIÓN DE PERFORMANCE
# ==============================================================================
print("\n" + "="*60)
print("STEP 3: OPTIMIZACIÓN Y ANÁLISIS DE CARTERA")
print("="*60)

def objective_function(weights, returns_history, expected_rets, lambda_risk=2.0, alpha=0.05):
    port_return = np.dot(weights, expected_rets)
    port_hist_rets = returns_history.dot(weights)
    var_threshold = np.percentile(port_hist_rets, alpha * 100)
    cvar = -port_hist_rets[port_hist_rets <= var_threshold].mean()
    return -(port_return - (lambda_risk * cvar))

n = len(returns.columns)
cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bounds = tuple((0.0, 0.25) for _ in range(n)) # Diversificación forzada max 25%
init_guess = [1/n] * n

opt_result = minimize(objective_function, init_guess, 
                      args=(returns, expected_returns, 2.0, 0.05), 
                      method='SLSQP', bounds=bounds, constraints=cons)

# Construir Carteras
w_naive = np.array([1/n]*n)
w_opt = opt_result.x

df_weights = pd.DataFrame({'Naive': w_naive, 'Optimized': w_opt}, index=returns.columns)
print("Pesos Finales (>1%):")
print(df_weights[df_weights['Optimized']>0.01].round(4))

# Cálculo de Historia
rets_naive = returns.dot(w_naive)
rets_opt = returns.dot(w_opt)
wealth_naive = 100 * (1 + rets_naive).cumprod()
wealth_opt = 100 * (1 + rets_opt).cumprod()

# --- GRÁFICA 4: RENTABILIDAD ACUMULADA ---
plt.figure(figsize=(12, 6))
plt.plot(wealth_naive, label='Benchmark Naive (Equal Weight)', color='gray', linestyle='--')
plt.plot(wealth_opt, label='Tu Cartera (IA + CVaR)', color='green', linewidth=2.5)
plt.title('La Carrera: Rentabilidad Acumulada (Base 100)', fontsize=16)
plt.legend()
plt.show()

# --- GRÁFICA 5: ROLLING VOLATILITY (RIESGO DINÁMICO) ---
# Muestra cómo tu cartera "sufre menos" en crisis
vol_rolling_naive = rets_naive.rolling(12).std() * np.sqrt(12)
vol_rolling_opt = rets_opt.rolling(12).std() * np.sqrt(12)

plt.figure(figsize=(12, 5))
plt.plot(vol_rolling_naive, label='Volatilidad Benchmark', color='red', alpha=0.5)
plt.plot(vol_rolling_opt, label='Volatilidad Tu Cartera', color='green', linewidth=2)
plt.title('Control de Riesgo: Volatilidad Móvil (12 meses)', fontsize=16)
plt.ylabel('Volatilidad Anualizada')
plt.legend()
plt.show()

# --- GRÁFICA 6: HISTOGRAMA DE RETORNOS (SEGURIDAD) ---
# Muestra que tu cartera tiene menos "días rojos" extremos
plt.figure(figsize=(12, 5))
plt.hist(rets_naive, bins=50, alpha=0.5, label='Benchmark', color='red', density=True)
plt.hist(rets_opt, bins=50, alpha=0.6, label='Optimized', color='green', density=True)
plt.title('Distribución de Retornos: Eliminando el Riesgo de Cola', fontsize=16)
plt.xlabel('Retorno Mensual')
plt.legend()
plt.show()

# Tabla Final de Métricas
def get_metrics(r):
    ann_ret = r.mean() * 12
    ann_vol = r.std() * np.sqrt(12)
    sharpe = (ann_ret - 0.02) / ann_vol
    max_dd = ((1+r).cumprod() / (1+r).cumprod().cummax() - 1).min()
    return [ann_ret, ann_vol, sharpe, max_dd]

metrics = pd.DataFrame({
    'Benchmark': get_metrics(rets_naive),
    'Optimized': get_metrics(rets_opt)
}, index=['Retorno Anual', 'Volatilidad', 'Sharpe Ratio', 'Max Drawdown'])

print("\n--- TABLA FINAL DE KPI ---")
print(metrics.round(4))
```
